## Intersubjectivity Collapse

This notebook presents a series of visualizations that illustrate the divergence of thought processes among a group of agents over time. Each agent represents an individual with its own unique perspective, modeled through random movements in a 2D space. The visualizations capture how these perspectives evolve and diverge, ultimately leading to a collapse in intersubjectivity.

By simulating the interactions of multiple agents, we can observe patterns of divergence that reflect real-world social dynamics. The visualizations highlight key moments where agents' paths diverge significantly, symbolizing breakdowns in shared understanding. This experiment serves as a tool for examining the effects of individual cognitive processes on collective experience and the potential consequences for social cohesion.


```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Sample data for biomass over billions of years
# For illustration purposes, we'll use a simple exponential decay function
years = np.linspace(0, 4.5, 100)  # 0 to 4.5 billion years
initial_biomass = 1000  # arbitrary units
decay_rate = 0.5  # arbitrary decay rate

# Biomass as a function of time (years)
biomass = initial_biomass * np.exp(-decay_rate * years)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(years, biomass, label='Biomass', color='green')
plt.xlabel('Time (billion years)')
plt.ylabel('Biomass (arbitrary units)')
plt.title('Biomass Over Billions of Years')
plt.legend()
plt.grid(True)
plt.show()
```

This code models the divergence of agents' thought processes over time and visualizes their evolving positions in 2D space. The animation illustrates how their perspectives shift, emphasizing the concept of intersubjective "collapse." It generates and validates the data frames before creating an interactive visualization to depict this phenomenon.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Example function to generate mock frames
def generate_mock_frames(num_frames=100, num_points=10):
    return [np.random.rand(num_points, 2) for _ in range(num_frames)]

# Validate data to ensure each frame has shape (N, 2)
def validate_frame(frame):
    if not isinstance(frame, np.ndarray):
        raise ValueError("Frame must be a numpy array.")
    if frame.ndim != 2 or frame.shape[1] != 2:
        raise ValueError(f"Frame must be of shape (N, 2). Got {frame.shape} instead.")

# Initialize frames
frames = generate_mock_frames()

# Validate all frames
for i, frame in enumerate(frames):
    try:
        validate_frame(frame)
    except ValueError as e:
        print(f"Validation error in frame {i}: {e}")
        raise

# Create the plot
fig, ax = plt.subplots()
scatter = ax.scatter([], [])

# Setting up the axes limits
ax.set_xlim(0, 1)  # Adjust to match your data range
ax.set_ylim(0, 1)  # Adjust to match your data range
ax.set_title("Scatter Plot Animation")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# Update function for animation
def update(frame):
    scatter.set_offsets(frame)  # Update scatter plot with new data
    return scatter,

# Create the animation
ani = FuncAnimation(
    fig, update, frames=frames, blit=True, repeat=True, interval=100
)

# Show the animation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_agents = 10  # Number of agents
steps = 100  # Number of time steps
divergence_factor = 0.1  # How quickly agents diverge

# Initialize agents in a 2D space
positions = np.random.rand(num_agents, 2)  # Random initial positions
trajectories = [positions.copy()]  # List to store trajectories

# Simulate divergence over time
for _ in range(steps):
    positions += np.random.randn(num_agents, 2) * divergence_factor  # Update positions with random walk
    trajectories.append(positions.copy())  # Store new positions

# Convert trajectories into arrays for plotting
trajectories = np.array(trajectories)

# Plot the trajectories
plt.figure(figsize=(10, 6))  # Set the figure size
for i in range(num_agents):
    plt.plot(trajectories[:, i, 0], trajectories[:, i, 1], label=f'Agent {i+1}')  # Plot each agent's trajectory

# Add plot titles and labels
plt.title("Visualization of Intersubjective Collapse")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize='small')  # Place legend outside the plot
plt.grid(True)  # Add grid for better readability
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()  # Display the plot
```

## Phantom Cognit

```python
import networkx as nx
import matplotlib.pyplot as plt

# Function to visualize the network
def visualize_network(G, title):
    pos = nx.spring_layout(G)  # Layout for visualization
    plt.figure(figsize=(10, 7))
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, edge_color='gray', font_size=15, font_weight='bold')
    plt.title(title)
    plt.show()

# Create a network (graph) with NetworkX
G = nx.Graph()

# Add nodes (cognitive units)
nodes = range(1, 11)  # Creating 10 nodes
G.add_nodes_from(nodes)

# Add edges (dependencies) - create a simple connected graph
edges = [(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (3, 7), (4, 8), (5, 8), (6, 9), (7, 10)]
G.add_edges_from(edges)

# Visualize the initial network
visualize_network(G, "Initial Network")

# Remove a node (cognit) to simulate the "missing cognit"
G.remove_node(3)

# Visualize the network after removing the node
visualize_network(G, "Network After Removing Node 3")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Create a cognitive network
nodes = 10  # Number of nodes
G = nx.erdos_renyi_graph(n=nodes, p=0.4)  # Random graph
pos = nx.spring_layout(G)  # Layout for visualization

# Plot original network
plt.figure(figsize=(10, 5))

# Original network
plt.subplot(1, 2, 1)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=10)
plt.title("Original Cognitive Network")

# Remove a node to simulate phantom cognit
G_removed = G.copy()
node_to_remove = np.random.choice(list(G.nodes))
G_removed.remove_node(node_to_remove)

# Plot network with missing cognit
plt.subplot(1, 2, 2)
nx.draw(G_removed, pos, with_labels=True, node_color='salmon', node_size=700, font_size=10)
plt.title(f"Network with Missing Cognit (Node {node_to_remove} Removed)")

plt.tight_layout()
plt.show()
```
